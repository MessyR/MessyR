```python
#1.Load the Iris dataset using Scikit Learn

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
iris = load_iris()
```

```python
#To see what's inside sklearn.datasets----->dir(sklearn.datasets)
```

```python
#Iris is basically a dictionary with the following keys

iris.keys()
```

```python
#Different feature names of Iris dataset

iris['feature_names']
```

```python
#2. Select only the Petal length and Petal width features

X = iris.data[:, 2:]  # selecting all rows and columns from 2 onwards i.e. petal length and width
y = iris.target
```

```python
#3. Train our Decision Tree classifier on the Iris Dataset

tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y) #when we call fit, decisiontreeclassifier is creating a tree with depth=2
```

```python
#4. Visualize our Decision Tree using export_graphviz()
#We can visualize the trained decision tree using the export_graphviz() method.graphviz is a great graph visualisation tool

from sklearn.tree import export_graphviz
export_graphviz( 
    tree_clf,
    out_file="iris_tree.dot", 
    feature_names=iris.feature_names[2:], 
    class_names=iris.target_names, 
    rounded=True,
    filled=True )
```
